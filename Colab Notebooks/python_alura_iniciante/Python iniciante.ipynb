# Aula 1 - Introdução ao Python

```python
nome = "Ana"
idade = 25
altura = 1.68
estudante = True

print("Nome:", nome)
print("Idade:", idade)
print("Altura:", altura)
print("É estudante?", estudante)

# Função para apresentar uma pessoa
def apresentar_pessoa(nome, idade, altura, estudante):
    print(f"Nome: {nome}")
    print(f"Idade: {idade}")
    print(f"Altura: {altura}")
    print(f"Estudante: {estudante}")

# Exemplo de uso da função
apresentar_pessoa("Carlos", 30, 1.75, False)
```

# Aula 2 - Operadores e Condicionais

```python
# Operadores matemáticos
soma = 10 + 5
subtracao = 10 - 5
multiplicacao = 10 * 5
divisao = 10 / 5
modulo = 10 % 3

print("Soma:", soma)
print("Subtração:", subtracao)
print("Multiplicação:", multiplicacao)
print("Divisão:", divisao)
print("Módulo:", modulo)

# Função de calculadora básica
def calculadora(a, b):
    print(f"{a} + {b} = {a+b}")
    print(f"{a} - {b} = {a-b}")
    print(f"{a} * {b} = {a*b}")
    print(f"{a} / {b} = {a/b}")
    print(f"{a} % {b} = {a%b}")

calculadora(15, 4)

# Condicionais
numero = 7

if numero > 0:
    print("O número é positivo.")
elif numero == 0:
    print("O número é zero.")
else:
    print("O número é negativo.")

# Função para verificar se é par ou ímpar
def par_ou_impar(numero):
    if numero % 2 == 0:
        print("Par")
    else:
        print("Ímpar")

par_ou_impar(10)
par_ou_impar(7)
```

# Aula 3 - Laços de Repetição

```python
# For loop
print("Contando até 5:")
for i in range(1, 6):
    print(i)

# While loop
contador = 1
print("Contando com while:")
while contador <= 5:
    print(contador)
    contador += 1

# Função para contar até um número
def contar_ate(numero_final):
    for i in range(1, numero_final + 1):
        print(i)

contar_ate(3)
```

# Aula 4 - Listas e Dicionários

```python
frutas = ["maçã", "banana", "laranja"]
print(frutas)
print(frutas[1])  # banana

# Adicionar item
frutas.append("uva")
print(frutas)

# Dicionários
aluno = {
    "nome": "João",
    "idade": 22,
    "curso": "Análise de Dados"
}

print(aluno["nome"])
print(aluno.get("curso"))

# Função para mostrar frutas
def mostrar_frutas(lista):
    for fruta in lista:
        print(fruta)

mostrar_frutas(frutas)
```

# Aula 5 - Funções e Bibliotecas

```python

# Importando biblioteca
import math

# Usando biblioteca
raiz = math.sqrt(16)
print("Raiz quadrada de 16:", raiz)

# Função que usa math
def calcular_raiz(numero):
    resultado = math.sqrt(numero)
    print(f"Raiz quadrada de {numero} é {resultado}")

calcular_raiz(25)

# Exemplo com random
import random

# Função para gerar número aleatório
def gerar_numero_aleatorio(inicio, fim):
    numero = random.randint(inicio, fim)
    print(f"Número aleatório entre {inicio} e {fim}: {numero}")

gerar_numero_aleatorio(1, 10)
```
