```python
from random import sample
```

```python

# Histórico real de sorteios
historico_sorteios = [
    [9, 24, 27, 36, 43, 54],
    [5, 8, 11, 15, 39, 47],
    [1, 3, 13, 16, 36, 56],
    [2, 12, 18, 21, 24, 37],
    [2, 9, 32, 38, 48, 55],
    [1, 28, 34, 36, 51, 52],
    [13, 22, 38, 46, 51, 56],
    [15, 21, 25, 55, 58, 59],
    [1, 11, 13, 27, 45, 59],
    [6, 10, 20, 54, 58, 60],
    [2, 38, 50, 54, 58, 59],
    [4, 16, 34, 43, 54, 57]
]
```

```python
# Criar uma lista com todos os números sorteados
lista = [num for sorteio in historico_sorteios for num in sorteio]
```

```python

# Criar um dicionário para contar a frequência de cada número
frequencias = {}
for num in lista:
    frequencias[num] = frequencias.get(num, 0) + 1  # Soma 1 a cada ocorrência
```

```python
# Contar a frequência e ordenar pelos mais sorteados
top_10 = sorted(set(lista), key=lista.count, reverse=True)[:10]
```

```python
# Dos top 10 escolher 6 aleatórios
numeros_sorteados = sample(top_10, 6)

print("Top 10 números mais sorteados:", top_10)
print("Números escolhidos:", sorted(numeros_sorteados))
```
