# **Plot functions**

```python
def plot_roc_curves(model, model_name, X_test, y_test):
    import matplotlib.pyplot as plt
    from sklearn.metrics import roc_curve, auc

    plt.figure(figsize=(10, 8))

    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_test)[:, 1]
    elif hasattr(model, "decision_function"):
        y_prob = model.decision_function(X_test)
    else:
        raise ValueError(f"O modelo {model_name} não suporta predict_proba ou decision_function.")

    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, lw=2, label=f'{model_name} (AUC = {roc_auc:.2f})')

    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')  # Linha diagonal (curva ROC aleatória)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Curva ROC para Múltiplos Modelos')
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.show()
```

```python
def evaluate_model(y_pred, y_test, r2_threshold=0.7, mae_threshold=10, mse_threshold=100):
    from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)

    is_good_model = r2 >= r2_threshold and mae <= mae_threshold and mse <= mse_threshold

    print(f"R²: {r2:.4f}")
    print(f"MAE: {mae:.4f}")
    print(f"MSE: {mse:.4f}")
    print(f"Modelo bom? {'Sim' if is_good_model else 'Não'}")
```

```python
def benchmark_auc(all_models, X, y):
    import numpy as np
    from sklearn.model_selection import cross_val_score
    from sklearn.model_selection import RepeatedStratifiedKFold
    for name, model in all_models:
      scores = []
      cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=5, random_state=1)
      scores.append(cross_val_score(model, X, y, scoring='roc_auc', cv=cv, n_jobs=-1))
      media =  np.mean(scores)
      print(f'Modelo: {name}')
      print(f'Média ROC_AUC : {media}')
      return np.mean(scores)
```

## Get hyperparameters

```python
def hyperparameter_search(model, param_grid, X_train, y_train, search_type='grid', cv=5, n_iter=50, scoring='accuracy', random_state=42, verbose=1):
    import numpy as np
    from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

    # param_grid: Dicionário com os hiperparâmetros a serem testados
    if search_type == 'grid':
        search = GridSearchCV(estimator=model, param_grid=param_grid, cv=cv, scoring=scoring, verbose=verbose)
    elif search_type == 'randomized':
        search = RandomizedSearchCV(estimator=model, param_distributions=param_grid, n_iter=n_iter, cv=cv, scoring=scoring, random_state=random_state, verbose=verbose)
    else:
        raise ValueError("search_type deve ser 'grid' ou 'randomized'")

    search.fit(X_train, y_train)

    best_model = search.best_estimator_
    best_params = search.best_params_
    best_score = search.best_score_

    print(f"Melhores Hiperparâmetros ({search_type}): {best_params}")
    print(f"Melhor Pontuação ({search_type}): {best_score}")

    return best_model, best_params, best_score
```

## Underfitting and Overfitting

```python
def detect_overfitting_and_underfitting(model, X_train, y_train, cv=5, threshold=0.03):
    from sklearn.model_selection import cross_val_score

    model.fit(X_train, y_train)
    train_score = model.score(X_train, y_train)

    val_scores = cross_val_score(model, X_train, y_train, cv=cv)
    val_score_mean = val_scores.mean()

    score_difference_ratio = (train_score - val_score_mean) / train_score

    print(f"Acurácia de Treinamento: {train_score:.4f}")
    print(f"Acurácia de Validação (média de {cv} folds): {val_score_mean:.4f}")
    print(f"Proporção de Diferença: {score_difference_ratio:.4f}")

    if train_score > val_score_mean and score_difference_ratio > threshold:
        print("Overfitting detectado!") # lembrando que pode ser algo aceitavel de acordo com o threshold
    elif val_score_mean >= train_score and abs(score_difference_ratio) < threshold:
        print("Underfitting detectado!")
    else:
        print("Modelo equilibrado, sem overfitting ou underfitting.")
```

## Outliers Detector

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(45)

df = pd.DataFrame({
    'Grupo': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Valores1': np.random.randint(10, 100, size=9),
    'Valores2': np.random.randint(15, 105, size=9),
    'Valores3': np.random.randint(5, 85, size=9)
})

# calcula a distância 'interquatilica' (IQR) entre N colunas que você passar (auto)
def calc_iqr(df, df_column: str) -> tuple:
  Q1 = df[df_column].quantile(0.25)
  Q3 = df[df_column].quantile(0.75)
  IQR = Q3 - Q1

  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  return lower_bound, upper_bound

def detect_and_return_outliers(df, df_columns):
  plt.figure(figsize=(10, 6))
  sns.boxplot(data=df[df_columns])
  plt.title('Boxplot para Detecção de Outliers')
  plt.show()

  outliers = []
  for column in df_columns:
      lower_bound, upper_bound = calc_iqr(df, column)
      _df = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
      if not _df.empty:
        for index in _df.index: outliers.append(int(index))
  return outliers
```
